# 함수 심화

## 함수 기본값

```python
def greet(name, greeting="환영합니다"):
    print(f"{name}님, {greeting}")

greet("대림대")
greet("대림대","안녕하세요")
```

```python
# 할인율 계산
# discount: 0.0~1.0 (0%~100%)
def calculate_price(price,discount=0):
    discountd = price * (1-discount)
    return discountd


print(f"기본 가격: {calculate_price(10000)}원")
print(f"할인 가격: {calculate_price(10000,0.1)}원")
```

## 함수에서 여러값 반환하기

```python
# 최솟값과 최대값을 동시 반환하는 함수
def min_max(numbers):
    minimum=min(numbers)
    maximum=max(numbers)
    return minimum,maximum
    
scores = [10,20,30,40,50,60,70,80,]
low, high =min_max(scores)
print("최솟값:", low)
print("최댓값:", high)
```

## 문제
class_scores = [45,62,78,55,90,83,71,66,48,95]
1. 성적 분석 함수 생성
2. 아래와 같은 로그 출력 - 4줄  
    - 평균: x.x 점 (소수점 1번째자리까지만 출력)
    - 최고점: xx점
    - 최저점:
    - 합격(60점 이상): x명

```python
def grade(score):
    pass_total=0
    for i in score:
        if i>=60:
            pass_total+=1
    return max(score),min(score),sum(score)/len(score),pass_total


class_scores = [45,62,78,55,90,83,71,66,48,95]

high,low,avg,hapguk=grade(class_scores)


print(f"평균: {avg:.1f}점")
print(f"최고점: {high}점")
print(f"최저점: {low}점")
print(f"합격(60점이상): {hapguk}명")
```

# Pandas 심화

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
# 학생 데이터
grade_data = {
    "name": ["kim","lee","park","cho","choi","jung"],
    "korean": [88,72,95,65,80,91],
    "math" :  [75,90,65,85,77,82],
    "english":[82,88,76,70,95,78]
}

grade_df=pd.DataFrame(grade_data)
grade_df
```

```python
print("--- 국어 점수 낮은 순 ---")
print(grade_df.sort_values("korean"))
print()
print("---- 수학 점수 높은 순")
print(grade_df.sort_values("math",ascending=False))
```

```python
grade_df["average"] = \
    (grade_df["korean"] + grade_df["math"] + grade_df["english"]) / 3

grade_df['average']= grade_df['average'].round(1)
grade_df
```

```python
grade_df["grade"] = grade_df["average"].apply(
    lambda x: "A" if x >= 80 else "B"
)

grade_df
```

## 그룹별 통계

```python
group_avg=grade_df.groupby("grade")["average"].mean()
print("등급별 평균:")
print(group_avg)
```

```python
# 응용 - 회사 부서별 급여 파악

employee_data = {
    "name" : ["kim","lee","park","cho","choi","jung"],
    "department" : ["개발", "마케팅", "개발", "인사", "마케팅", "인사"],
    "salary" : [4200, 3500 ,4800 ,3200 ,3800, 5000]
}

emp_df=pd.DataFrame(employee_data)


# 부서별 평균 급여

dept_avg=emp_df.groupby("department")["salary"].mean()
print("---- 부서별 평균 급여 ----")
print(dept_avg.round(0))
print()
print("---- 부서별 인원 수 ----")
dept_count=emp_df.groupby("department")["name"].count()
print(dept_count)
```
